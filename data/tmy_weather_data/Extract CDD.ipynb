```python
import os
import pandas as pd
import csv
import numpy as np
```

```python
for fn in os.listdir('data'):
    path = os.path.join('data', fn)
    with open(path) as f:
        for row in csv.reader(f):
            metadata = row
            break
        data = pd.read_csv(f)
    temperatureC = np.reshape(data['Dry-bulb (C)'], (365, 24))
    temperatureF = temperatureC * 1.8 + 32
    
    cdd = np.sum(temperatureF[temperatureF > 65] - 65) / 24
    hdd = np.sum(65 - temperatureF[temperatureF < 65]) / 24
    print(hdd)
    print(cdd)
    break
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(np.reshape(temperatureF, (365, 24)).T)
plt.show()
```
